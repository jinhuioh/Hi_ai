9. knn은 어떤 모델인가? 유클리디안 거리와 맨하탄 거리를 비교?
K-최근접 이웃(K-Nearest Neighbor, KNN) : 
분류할 때 홀수로 거리상으로 가장 가까운 데이터들을 찾아서 다수결로 어떤 값인지 분류하는 방법
거리를 잴때 직선으로 구하면 유클리디안거리, 계단식으로 구하면 멘하탄거리
유클리디안거리:  루트(가로^2 + 세로^2) 로 직선의 거리이고, 
멘하탄 거리:      모눈종이의 연속적인 정사각형의 가로세로 길이를
                       합한것으로, 직선의 거리가 아니다.
