```python
import folium
```

```python
#화면에 지도를 띄우자
map = folium.Map(location=[37.57, 126.98],tiles='stamen Toner',zoom_start=13)#tiles='stamen Toner' 지도 모양 변경 zoom_start=13은 줌 되는 정도.
map
```

```python
#화면에 지도를 띄우자
map = folium.Map(location=[37.57, 126.98],zoom_start=20)#tiles='stamen Toner' 지도 모양 변경 zoom_start=13은 줌 되는 정도.
map
```

```python
#마커를 찍어보자
map = folium.Map(location=[37.57, 126.98], zoom_start=20)#tiles='stamen Toner' 지도 모양 변경 zoom_start=13은 줌 되는 정도.
folium.Marker([37.57, 126.98],popup='여기는 종각역!',#폴리움 마커 객체 생성 #popup='여기는 종각역!' 팝업창 글씨 쓸 수 있음
              icon=folium.Icon(color='red', icon='info-sign')).add_to(map)#add_to(map)하면 맵 안에 객체 넣음
map
```
