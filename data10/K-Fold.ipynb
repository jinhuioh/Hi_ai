```python
titanic_df = pd.read_csv('./titanic_train.csv')
(titanic_df.head())
```

```python
titanic_df.info()
```

```python
##함수(drop)
##name,ticket,body,home.dest
def drop_features(df):#타이타닉 df를 넣고
    df.drop(columns=['name','ticket','body','home.dest'],inplace=True)
    return df#결과값을 리턴
```

```python
titanic_df =  drop_features(titanic_df)
titanic_df.head()
```

```python
titanic_df.info()
```

```python
##na처리
##age는 평균으로 채워주세요
##cabin은 N으로 채워주세요
##embarked는 N으로 채워주세요(fillna())
##na개수가 몇개인지 전체 카운트
```

```python
def fillna(df):
    df['age'].fillna(df['age'].mean(), inplace=True)
    df['cabin'].fillna('N',inplace=True)    
    df['embarked'].fillna('N',inplace=True)     
    return df
```

```python
titanic_df = fillna(titanic_df)
titanic_df
```

```python
titanic_df.info()
```

```python
#라벨인코딩:sex,cabin,embarked
```

```python
titanic_df['cabin'].head(10)
```

```python
#cabin컬럼의 첫글자만 넣는걸로 바꾸기, 라벨링 인코딩 하기 위해
# df['cabin'] = df['cabin'].str[:1]
# df['cabin'].head(10)
```

```python
#라벨링 함수화
def label_encoding(df):
    df['cabin'] = df['cabin'].str[:1]
    print(list(titanic_df['cabin'])[:10])    
    le = LabelEncoder()#라벨인코딩하는 객체 생성
    features = ['sex','cabin','embarked']
    for f in features:
        le.fit(df[f])
        print('라벨링할 목록=>', le.classes_)
        df[f] =le.transform(df[f])
    return df
```

```python
titanic_df = label_encoding(titanic_df)
titanic_df
```

```python
titanic_df.head()
```

```python
#지금까지 만든함수 한번에 묶어서 불러주기.
def preprocessing(df):
    df2 = fillna(df)
    df3 = drop_features(df2)
    df4 = label_encoding(df3)
    return df4
```

```python
titanic_df2 = pd.read_csv('./titanic_train.csv')
(titanic_df2.head())
```

```python
#데이터 전처리 호출
titanic_df3 = preprocessing(titanic_df2)
titanic_df3
```
