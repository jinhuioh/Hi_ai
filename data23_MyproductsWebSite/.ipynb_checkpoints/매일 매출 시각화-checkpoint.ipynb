```python
df = pd.read_excel("./Statistics-Daily-20220103_20220701_(0).xlsx")
df_month = df.iloc[149:179]
df_month
```

```python
df_all_price= df_month['전체 거래 금액']
df_all_price
```

```python
df_date = df_month['날짜']
df_date
```

```python
fig = plt.figure(figsize=(50,50))#캔버스 생성
fig.set_facecolor('white')# 캔버스 색상 설정
ax = fig.add_subplot()#그림 뼈대 생성

ax.plot(df_date, df_all_price, color='blue')#선그래프 생성
plt.rc('font', size=50)
# plt.xticks(rotation=45)
plt.title('매출금액',fontsize=50)
plt.show()
```
